```python
# Import dependencies
import pandas as pd

# Read in csv from previous project containing all counties in the US by their unique GEO_ID
counties = pd.read_csv('../cleaned_data/us_counties.csv')

# Read in csv containing every county in the US and the coordinates of its midpoint.
midpoints = pd.read_csv('../cleaned_data/us_county_midpoints.csv', dtype={'GEOID': str})
```

```python
counties.head()
```

```python
midpoints.head()
```

```python
# Merge the columns on their unique Geo_ID columns
merged = midpoints.merge(counties, how="outer",on="Geo_ID")
merged.head()
```

```python
# Drop duplicate and unncessary columns
merged = merged.drop(columns=["County_y","State_Abbrev_y","Geo_ID", "County_x", "State_Abbrev_x","State"])
merged.head()
```

```python
# Fix names of duplicated columns
merged.rename(columns={'County_x': 'County', 'State_Abbrev_x': 'State_Abbrev'}, inplace=True)
merged.head()
```

```python
# Export final csv file
merged.to_csv('../cleaned_data/corrected_midpoints.csv', index=False)
```
