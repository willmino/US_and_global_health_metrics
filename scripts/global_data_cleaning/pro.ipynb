```python
gdp_per_capita = pd.read_csv('raw_data/gdp_per_capita_worldbank.csv')
gdp_per_capita=gdp_per_capita.loc[(gdp_per_capita["Year"]==2015)|(gdp_per_capita["Year"]==2016)|(gdp_per_capita["Year"]==2017)|(gdp_per_capita["Year"]==2018)|(gdp_per_capita["Year"]==2019)]
gdp_per_capita=gdp_per_capita.rename(columns={"GDP per capita, PPP (constant 2017 international $)":"gdp_per_capita"})
gdp_per_capita=gdp_per_capita.rename(columns={'Entity':'country'})
gdp_per_capita["country"]=gdp_per_capita["country"].str.replace('[(].+[)]','')
gdp_per_capita["country"]=gdp_per_capita["country"].str.replace('\s$','')
# Save the modified dataframe to a new file
gdp_per_capita.to_csv('modified_file.csv', index=False)
gdp_per_capita=gdp_per_capita.groupby("country").mean()[["gdp_per_capita"]]
gdp_per_capita["gdp_per_capita"]=round(gdp_per_capita["gdp_per_capita"]).astype('int')
gdp_per_capita=gdp_per_capita.reset_index()
gdp_per_capita['country_code'] = gdp_per_capita['country'].map(master_country_dict)
# Save the modified dataframe to a new file
gdp_per_capita.to_csv('modified_file.csv', index=False)
gdp_per_capita=gdp_per_capita.groupby("country").mean()[["gdp_per_capita"]]
gdp_per_capita["gdp_per_capita"]=round(gdp_per_capita["gdp_per_capita"]).astype('int')
gdp_per_capita=gdp_per_capita.reset_index()
gdp_per_capita['country_code'] = gdp_per_capita['country'].map(master_country_dict)
gdp_per_capita=gdp_per_capita.loc[gdp_per_capita["country"]!='World']
gdp_per_capita=gdp_per_capita[~gdp_per_capita['country_code'].isnull()]
gdp_per_capita=gdp_per_capita.drop(columns="country")
gdp_per_capita
```

```python
import requests
import pandas as pd
import time

country_codez=[]
countries=[]
cities=[]
latitudes=[]
longitudes=[]
populations=[]

#create an empty dataframe to store the results
df = pd.DataFrame()

# loop through each country code and retrieve the largest cities by population
for country_code in country_codes:
    try:
        # make an API call to the GeoDB Cities API to retrieve information on the largest cities by population
        response = requests.get(f"https://wft-geo-db.p.rapidapi.com/v1/geo/cities?countryIds={country_code}&minPopulation=10000&limit=10&sort=-population", headers={"X-RapidAPI-Key": GeoDB_API_KEY})
        response.raise_for_status()  # raise an exception for 4xx or 5xx HTTP status codes
        
        json = response.json()["data"]
        
        # create a set to keep track of unique cities for this country
        unique_cities = set()
        
        # start the timer
        start_time = time.time()

        for i in json:
            # only add the city to the list if it hasn't already been added
            if i["city"] not in unique_cities:
                country_code=i["countryCode"]
                country=i["country"]
                city = i["city"]
                latitude=i["latitude"]
                longitude=i["longitude"]
                population=i["population"]
                country_codez.append(country_code)
                countries.append(country)
                cities.append(city)
                latitudes.append(latitude)
                longitudes.append(longitude)
                populations.append(population)
                unique_cities.add(city)

            # if we already have 10 unique cities for this country or the loop has been running for more than 60 seconds, break out of the loop
            if len(unique_cities) == 10 or time.time() - start_time > 60:
                break
        
        # Pause the execution of the code for 10 seconds before making the next API request
        time.sleep(1)
    except requests.exceptions.HTTPError as e:
        print(f"Error from {country_code}: {e}")
        time.sleep(1)
        continue

df["country_code"]=country_codez
df["country"]=countries
df["city"]=cities
df["latitude"]=latitudes
df["longitude"]=longitudes
df["population"]=populations

# save the results to a CSV file
df.to_csv("city_coordinates_unique.csv", index=False)
```

```python
for i, row in city_coordinates_part3.iterrows():
    try:
        # make an API call to Openweathermap Historical Air Pollution API to retrieve information on the largest cities by population
        response = requests.get(f'http://api.openweathermap.org/data/2.5/air_pollution/history?lat={row["latitude"]}&lon={row["longitude"]}&start=1640995200&end=1656633600&appid={API_KEY}')
        response.raise_for_status()  # raise an exception for 4xx or 5xx HTTP status codes

        json = response.json()["list"]

        co_list=[]
        no_list=[]
        no2_list=[]
        o3_list=[]
        so2_list=[]
        pm2_5_list=[]
        pm10_list=[]
        nh3_list=[]

        for j in json:
            co=j["components"]["co"]
            no=j["components"]["no"]
            no2=j["components"]["no2"]
            o3=j["components"]["o3"]
            so2=j["components"]["so2"]
            pm2_5=j["components"]["pm2_5"]
            pm10=j["components"]["pm10"]
            nh3=j["components"]["nh3"]
            
            co_list.append(co)
            no_list.append(no)
            no2_list.append(no2)
            o3_list.append(o3)
            so2_list.append(so2)
            pm2_5_list.append(pm2_5)
            pm10_list.append(pm10)
            nh3_list.append(nh3)

        city_coordinates_part3.at[i, "co"] = sum(co_list)/len(co_list)
        city_coordinates_part3.at[i, "no"] = sum(no_list)/len(no_list)
        city_coordinates_part3.at[i, "no2"] = sum(no2_list)/len(no2_list)
        city_coordinates_part3.at[i, "o3"] = sum(o3_list)/len(o3_list)
        city_coordinates_part3.at[i, "so2"] = sum(so2_list)/len(so2_list)
        city_coordinates_part3.at[i, "pm2_5"] = sum(pm2_5_list)/len(pm2_5_list)
        city_coordinates_part3.at[i, "pm10"] = sum(pm10_list)/len(pm10_list)
        city_coordinates_part3.at[i, "nh3"] = sum(nh3_list)/len(nh3_list)
        print(row["country"]+"_"+row["city"])
        time.sleep(1)

    except requests.exceptions.HTTPError as e:
        owmhapapi_country=row["country"]
        print(f"Error from {owmhapapi_country}: {e}")
        time.sleep(1)
        continue

    except ZeroDivisionError:
        print("Division by zero error occurred. Skipping City.")
        time.sleep(1)
        continue
city_coordinates_part3.to_csv('complete_pollution_data_part2.csv')
```
