```python
import pandas as pd
import numpy as np
import re
```

```python
cancer_list=["bladder","brain","breast","breast_insitu","cervix","colon","esophagus","kidney and renal","leukemia",
            "liver","lung","melanoma","non-hudgkin lymphoma","oral cavity","ovay","pancreas","prostate","stomach","thyroid",
             "uteras"]
```

```python
len(cancer_list)
```

```python
data1 = pd.read_csv("incd.csv",skiprows=8,skipfooter=36)#,index_col=1)
#data1.sort_values("State",ascending=True,inplace=True)
data1.head()
```

```python
data1.columns
```

```python
data1.rename(columns={" FIPS":"FIPS"},inplace=True)
data = data1[["FIPS","State","Age-Adjusted Incidence Rate([rate note]) - cases per 100,000","Lower 95% Confidence Interval","Upper 95% Confidence Interval"]]
data.head()
```

```python
data.dtypes
```

```python
data.rename(columns={"Age-Adjusted Incidence Rate([rate note]) - cases per 100,000":"Rate","Lower 95% Confidence Interval":"lower","Upper 95% Confidence Interval":"upper"},inplace=True)
data.head()
```

```python
data.columns
```

```python
for (ind,item) in enumerate(cancer_list): 
    name = str("incd ("+ str(ind+1)+").csv") 
    df = pd.read_csv(name,skiprows=8,skipfooter=36,engine='python')#,index_col=1,engine='python')
    df.rename(columns={" FIPS":"FIPS"},inplace=True)
    df = df[["FIPS","State","Age-Adjusted Incidence Rate([rate note]) - cases per 100,000","Lower 95% Confidence Interval","Upper 95% Confidence Interval"]]
    df.rename(columns={"Age-Adjusted Incidence Rate([rate note]) - cases per 100,000":item,"Lower 95% Confidence Interval":str("lower"+item),
                       "Upper 95% Confidence Interval":str("upper"+item)},inplace=True)
    data = data.merge(df)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.replace(to_replace=r'.*\*\.*',value=0,inplace=True,regex=True)
#data.replace(to_replace=r'.*[data not found].*',value=np.nan,inplace=True,regex=True)
#data.replace(to_replace=r'.*[data not available].*',value=np.nan,inplace=True,regex=True)
#data.replace(to_replace=r'.*\N\/A\.*',value=np.nan,inplace=True,regex=True)
#data.replace(to_replace=r'.*\3 or fewer\.*',value=np.nan,inplace=True,regex=True)
```

```python
data.head()
```

```python
def clean_c(column):
    for i in range(len(data[column])):
        if ((str(data[column].iloc[i]).strip() == "data not available")
           or (str(data[column].iloc[i]).strip() == "N/A")):
            data[column].iloc[i] = np.nan
        elif (str(data[column].iloc[i]).strip() == "3 or fewer"):
            data[column].iloc[i] = 3.0
    data[column] = data[column].astype("float")
```

```python
for i in data.columns[2:]:
    clean_c(i)
```

```python
data.head()
```

```python
data["Rate_range"] = data["upper"] - data["lower"]
data.drop(columns=["upper","lower"],inplace=True)
```

```python
for i in cancer_list:
    lower = str("lower"+i)
    upper = str("upper"+i)
    name = str(i + "_range")
    data[name] = data[upper] - data[lower]
    data.drop(columns=[upper,lower],inplace=True)
```

```python
data.head()
```

```python
data["State"] = data["State"].replace(to_replace=r"\([0-9]\)",value="",regex=True)
```

```python
data.head()
```

```python
data.to_csv("cancer_per_county.csv",index=False,na_rep="NaN")
```
